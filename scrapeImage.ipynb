```python
link = 'https://storage.googleapis.com/openimages/web/visualizer/index.html?set=train&type=segmentation&r=false&c=%2Fm%2F0bt_c3'
base_url = "https://storage.googleapis.com"
browser = webdriver.Chrome(r"C:\Users\sydel\Downloads\chromedriver_win32_83\chromedriver.exe")
# browser = webdriver.Chrome()

# Tell Selenium to get the URL you're interested in.
browser.get(link)

# Selenium script to scroll to the bottom, wait 3 seconds for the next batch of data to load, then continue scrolling.  It will continue to do this until the page stops loading new data.
lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
while(match==False):
    lastCount = lenOfPage
    time.sleep(3)
    lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    if lastCount==lenOfPage:
        match=True

# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source

# Throw your source into BeautifulSoup and start parsing!
bs_data = BeautifulSoup(source_data)
#------------------
```

```python
# print(bs_data)
image = bs_data.find_all('image')
image_link = []
for img in image:
    if img.get('data-src') is not None:
        image_link.append(img.get('data-src'))
# print(image)
print(image_link)
```

```python
bash_script = "image=("
for i in image_link:
    bash_script += i +" "
print(bash_script+")")
```

i wanted more images than the number of links listed when the google page first loads so i decided on an arbitary number of pages that i would scroll down and retrieve all the image links that were present in the page after the scrolls down. This is done by adding the variable count to the while loop.

```python
link = 'https://www.google.com/search?q=canned+food&client=firefox-b-d&sxsrf=ALeKk00mKhIMU3BM75QpuvixYFvO6z24Gw:1594395578497&source=lnms&tbm=isch&sa=X&ved=2ahUKEwiax9mwgsPqAhUVheYKHQ5tCagQ_AUoAXoECA4QAw&biw=1280&bih=606'
base_url = "https://www.google.com"
browser = webdriver.Chrome(r"C:\Users\sydel\Downloads\chromedriver_win32_83\chromedriver.exe")
# browser = webdriver.Chrome()

# Tell Selenium to get the URL you're interested in.
browser.get(link)

# Selenium script to scroll to the bottom, wait 3 seconds for the next batch of data to load, then continue scrolling.  It will continue to do this until the page stops loading new data.
lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
count=0
while(match==False and count<15):
    lastCount = lenOfPage
    time.sleep(3)
    lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    count+=1
    if lastCount==lenOfPage:
        match=True

# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source

# Throw your source into BeautifulSoup and start parsing!
bs_data = BeautifulSoup(source_data)
```

```python
canImage = bs_data.find_all('img')
canImageLink=[]
for img in canImage:
    if img.get('src') is not None and img.get('src').find('https')!=-1:
        canImageLink.append(img.get('src'))
print(canImageLink)
print(len(canImageLink))
# print(canImage)
```

```python
bash_script = "image=("
for i in canImageLink:
    bash_script += "'"+i + "'"+" "
print(bash_script+")")
```
